```python
import pandas as pd
```

```python
# 读取 Parquet 文件
df = pd.read_parquet('../../data/wsdm-cup/train.parquet')
```

```python
# 查看数据的前几行
print(df.head())
```

```python
print(df)
```

```python
# 打印所有列名
print(df.columns)
```

```python
# 获取指定字段的唯一值
unique_values = df['language'].unique()

# 输出唯一值
print(unique_values)
```

```python
# 筛选 'language' 为 'unknown' 的行，并只显示 'prompt' 和 'language' 列
unknown_language_df = df[df['language'].str.strip().str.lower() == 'unknown'][['prompt', 'language']]

# 输出筛选结果
print(unknown_language_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 统计所有语言出现的次数，并去除出现次数为 0 的语言
language_counts = df['language'].value_counts()

# 去除出现次数为 0 的语言（如果有的话）
language_counts = language_counts[language_counts > 10]

# 使用 matplotlib 或 seaborn 绘制柱状图
plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.index, y=language_counts.values, palette="viridis")

# 设置标题和标签
plt.title('Distribution of Languages in the Dataset (Excluding ten Counts)')
plt.xlabel('Language')
plt.ylabel('Count')

# 显示图表
plt.xticks(rotation=45)  # 旋转x轴标签以防止重叠
plt.show()
```
